```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python

home_prices = pd.read_csv(
                "data/homeprices.csv"
            )
```

```python
reg_model = linear_model.LinearRegression()
estmtr = reg_model.fit(home_prices[["area"]].values, home_prices.price.values)
```

```python
areas = pd.Series(
    [
        1000,
        1500,
        2300,
        3540,
        4120,
        4560,
        5490,
        3460,
        4750,
        2300,
        9000,
        8600,
        7100,
    ]
)

rstate_data = pd.DataFrame()
rstate_data["Areas"] = areas
```

```python
plt.figure(figsize = (15, 10))
plt.scatter(home_prices.area, home_prices.price, marker='*', color = 'red')
plt.plot(rstate_data.Areas, reg_model.predict(rstate_data[["Areas"]].values))
plt.show()
```
